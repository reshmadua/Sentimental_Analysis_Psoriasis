## 2. Extracting all tweets on Psoriasis for last 4-years(Approx)

```python
hashtag = ["Psoriasis"]
Limit = 100000
last_24_date_time = datetime.datetime.now() - datetime.timedelta(days = 1500)
today = last_24_date_time.strftime('%Y-%m-%d')
date = str(today)
c = twint.Config()
c.Lang = "en"
c.Search = hashtag
c.Limit = Limit
c.Since = date
c.Store_csv = True
c.Pandas = True
c.Pandas_clean = True
#c.Min_likes = 20
c.Output = "Psoriasis_3.csv"
twint.run.Search(c)
# data = twint.storage.panda.Tweets_df

#append_data.append(data)

#print('Searching: ' + hashtag)

time.sleep(10)
```
